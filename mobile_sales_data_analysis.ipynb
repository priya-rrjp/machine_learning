# **DATA DICTIONARY**



**brand**

- categorical(Nominal)

- Brand of the phone

**model**

- categorical(Nominal)

- model of the phone

**base_color**

- categorical(Nominal)

- base color of the phone

**processor**

- categorical(ordinal)

- processor type of the phone

**screen_size**

- categorical(ordinal)

- screen size of the phone

**ROM**

- numerical(discrete)

- The storage capacity in GB

**RAM**

- numerical(discrete)

- The RAM capacity in GB

**display_size**

- numerical(continous)

- The size of the phone's display in inches

**num_rear_camera**

- numerical(discrete)

- The number of rear cameras

**num_front_camera**

- numerical(discrete)
-  The number of front cameras

**battery_capacity**

- numerical(discrete)

- The number of front cameras

**ratings**

- numerical(continous)
- The average user ratings

**num_of_ratings**
- numerical(discrete)
- The number of ratings

**sales_price**
- numerical(discrete)
- The sales price of the phone

**discount_percent**
- numerical(continous)
- The discount percentage

**sale**
- numerical(continous)
- The number of units sold
